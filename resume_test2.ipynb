```python
from parsing_module import extract_text_pdfplumber, split_resume_into_sections
from experience_splitter import split_experience_section, parse_job_entry

# === Step 1: Load Resume and Extract Sections ===
pdf_path = "docs/sample_resume.pdf"
resume_text = extract_text_pdfplumber(pdf_path)
sections = split_resume_into_sections(resume_text)

# === Step 2: Pull Experience Section ===
experience_text = sections.get("experience", "").strip()
if not experience_text:
    print("[Fallback] No experience section found. Using full resume text.")
    experience_text = resume_text

# === Step 3: Split into Chunks and Parse ===
chunks = split_experience_section(experience_text)

for idx, chunk in enumerate(chunks):
    print(f"\n=== Raw Job Chunk {idx+1} ===\n")
    print(chunk[:600])  # preview first 600 chars

    job = parse_job_entry(chunk)
    print("\n→ Parsed Job Structure:")
    print(f"Company: {job['company']}")
    print(f"Title: {job['title']}")
    print(f"Date Range: {job['date_range']}")
    print("Bullets:")
    for b in job["bullets"]:
        print("  -", b.strip())
    print("="*60)
```
